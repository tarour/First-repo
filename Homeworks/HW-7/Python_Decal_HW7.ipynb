```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

twtdata = pd.read_json('AOC_recent_tweets.txt')

display(twtdata)
```

```python
display(twtdata.sort_values('id'))
```

```python
print(twtdata['created_at'])
dechourvalues = []
for i in twtdata['created_at']:
    j = str(i)
    l, lis = j.split(' ')
    hms, r = lis.split('+')
    h, m, s = hms.split(':')
    h, m, s = float(h), float(m), float(s)
    dechour = h + (m / 60) + (s / 3600)
    dechourvalues.append(dechour)

twtdata['hours'] = dechourvalues
display(twtdata)
```

```python
newtwtdata = pd.DataFrame({'created_at': twtdata['created_at'], 'hours': twtdata['hours'],'full_text':twtdata['full_text']})
display(newtwtdata)
```

```python
newtwtdata.to_csv('AOC_Tweets_HW7.csv')
```

```python
planetdata = sns.load_dataset('planets')
display(planetdata)
```

```python
methods = np.unique(planetdata['method'], return_counts=True)
print(methods)

totalist = []

for j in methods[0]:

    jlist = []
#    jlist.append(j)
    for i in range(len(planetdata['method'])):

        if planetdata['method'][i] == j:
            jlist.append(i)
    
    totalist.append(jlist)

colors = ['b','g','r','c','m','y','k','pink','orangered','aqua']
labels = ['Astrometry','Eclipse Timing Variations','Imaging','Microlensing','Orbital Brightness Modulation','Pulsar Timing','Pulsation Timing Variations','Radial Velocity','Transit','Transit Timing Variations']
print(totalist)
print(len(totalist))
```

```python
plt.figure(0)
for i in range(10):
    xvs = []
    for j in range(len(totalist[i])):
        xvs.append(totalist[i][j])

    print(totalist[i])
    print(planetdata['orbital_period'][xvs])
    plt.scatter(planetdata['orbital_period'][xvs],planetdata['mass'][xvs],c=colors[i],label=labels[i])
    plt.legend(loc='best',framealpha=.25)
plt.yscale('log')
plt.xscale('log')
plt.show()
```

```python
# Obtain counts for each year (y), and make it corresponding to the unique values of the column years (x)

yearcounts = np.unique(planetdata['year'], return_index=True, return_counts=True)
years, indices, counts = yearcounts
totalisty = []

for j in years:

    jlist = []
#    jlist.append(j)
    for i in range(len(planetdata['year'])):

        if planetdata['year'][i] == j:
            jlist.append(i)
    
    totalisty.append(jlist)
print(years)
print(totalisty)
```

```python
ovrlist = []
for k in totalist:
    jlist = []
    for j in years:
        
        counts = 0
        for i in k:
        
            if planetdata['year'][i] == j:

                counts += 1
        jlist.append(counts)    

    ovrlist.append(jlist)    
print(ovrlist)
print(len(years))
```

```python
barplot = pd.DataFrame({labels[0]: ovrlist[0],labels[1]: ovrlist[1],labels[2]: ovrlist[2],labels[3]: ovrlist[3],labels[4]: ovrlist[4],labels[5]: ovrlist[5],labels[6]: ovrlist[6],labels[7]: ovrlist[7],labels[8]: ovrlist[8],labels[9]: ovrlist[9]}, index = years)
ax = barplot.plot.bar(stacked=True)
```
